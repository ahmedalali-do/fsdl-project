# Fortune 500 Banks App Reviews Collection

The data is collected for 24 banks from both Apple App Store & Google Play Store

We will start with the App Store

```python
#import required libraries
import pandas as pd
from app_store_scraper import AppStore
from pprint import pprint
```

```python
bk = pd.read_csv('F500_Bank_Apps - Sheet1.csv')
bk.dropna(inplace=True)
bk['App Store ID'] = bk['App Store ID'].astype(int)

print(bk.shape)
bk.head()
```

```python
import datetime

date = datetime.datetime.strptime('2019-01-01', '%Y-%m-%d')
```

```python
for i in range(len(bk['App Name'])):

    x = AppStore(country=bk['Country'].iloc[i], app_name=bk['App Name'].iloc[i],
                                              app_id=bk['App Store ID'].iloc[i])

    x.review(after = date, sleep=10)

    z = pd.DataFrame(x.reviews)
    z['Store'] = 'App Store'
    z['App Name'] = bk['App Name'].iloc[i]

    if i == 0:

        df = pd.DataFrame(x.reviews)
        df['Store'] = 'App Store'
        df['App Name'] = bk['App Name'].iloc[i]

    else:

        df = df.append(z)
```

```python
df.to_csv('Mobile_Banking_App_Reviews2.csv')
```
